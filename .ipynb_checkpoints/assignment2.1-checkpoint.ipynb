```python
import numpy as np
import pandas as pd
from keras import layers, models
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_json('train.json')
test = pd.read_json('test.json')
```

Remove stopwords and format for vectorization

```python
def format_recipes(recipes):
    stopwords = ['water', 'boiling', 'cold', 'warm', 'cooking spray', 'salt', 'salted', 'unsalted', 'frozen', 'organic', 
                 'prepared', 'canned', 'fresh', 'dried', 'freshly', 'leaves', 'fat', 'free', 'less', 'low', 'reduced', 'sodium',
                 'ripe', 'cracked', 'ground', 'extra virgin', 'extra-virgin', 'seedless', 'mix', 'chopped', 'grated',
                 'shredded', 'large', 'medium', 'small', 'granulated', 'oz', 'lb', 'allpurpose', 'all purpose', 'all-purpose', 'all', 'and', 'any', 'applewood smoked', 
 'aâ',
 'aã',
                ]
    
    #  '10oz', '14', '14oz', '15oz', '1inchthick', '21', '25', '2to3lb', '33', '40', '5oz', '95', '96', '8ounc',
    
    result = []
    for recipe in recipes:    
        for i in range(len(recipe)):
            ingredient = recipe[i].split(' ')
            for j in range(len(ingredient) - 1, -1, -1):
                if ingredient[j] in stopwords or ingredient[j].isdigit():
                    del ingredient[j]
                if ingredient[j][-1] == 's':
                    ingredient[j] = ingredient[:-1]
            recipe[i] = ''.join(ingredient)
        recipe = ' '.join([word for word in recipe if word != ''])
        result.append(recipe)
    return result
```

```python
train['ing_list'] = format_recipes(train['ingredients'])
test['ing_list'] = format_recipes(test['ingredients'])
```

```python
cv = CountVectorizer()
Xtrain = cv.fit_transform(train['ing_list'])
Xtest = cv.transform(test['ing_list'])
```

```python
def to_one_hot(labels, dimension=20):
    results = np.zeros((len(labels), dimension))
    for i, label in enumerate(labels):
        results[i, label] = 1.
    return results
```
